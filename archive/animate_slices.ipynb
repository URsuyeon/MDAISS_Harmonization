```python
# -*- coding: utf-8 -*-
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc

rc('animation', html='jshtml')
```

```python
def animate_slices(frames, fps=5):
    fig = plt.figure(figsize=(6, 6))
    plt.axis('off')
    im = plt.imshow(frames[0], cmap='gray', origin='lower')
    
    def update(i):
        im.set_array(frames[i])
        return [im]
    
    return animation.FuncAnimation(
        fig, update, frames=len(frames),
        interval=1000//fps, blit=True
    )
```

```python
nifti_path = "/public/sylee/MDAISS_2/output/nifit/EUMC_S3_00014_S3_00014_TOF_grayscale.nii.gz"
img = nib.load(nifti_path)
vol = img.get_fdata()

print("Shape:", vol.shape)
print("Voxel size (Resolution):", img.header.get_zooms())
print("Affine matrix:\n", img.affine)

axis = 2
num_slices = vol.shape[axis]

frames = [np.take(vol, idx, axis=axis) for idx in range(num_slices)]

frames = [f for f in frames if f.max() > 0]
print(f"Animating {len(frames)} non-empty slices over axis {axis}")

anim = animate_slices(frames, fps=10)
anim
```
